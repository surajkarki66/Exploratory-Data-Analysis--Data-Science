## Step 4: Data Preperation
Before start to anaylze the data, we have to make sure that the data is cleaned. Following steps will be performed during data preparation.
- Dropping irrelevant columns and rows
- Identifying duplicated columns
- Renaming Columns
- Feature Creation

```python
# Example of dropping columns
# df.drop(['Opening date'], axis=1)
```

```python
# removing columns
df = df[[#'id',
         'title', 'description', 'publishTime', 
         #'kind_stats',
       'duration_seconds', 'viewCount', 'likeCount', 'commentCount',
       #'thumbnails.default.url', 'thumbnails.default.width',
       #'thumbnails.default.height', 'thumbnails.medium.url',
       #'thumbnails.medium.width', 'thumbnails.medium.height',
       #'thumbnails.high.url', 'thumbnails.high.width',
       #'thumbnails.high.height', 'contentDetails.duration',
       #'contentDetails.dimension', 'topicDetails.topicCategories',
       #'snippet.defaultLanguage', 'localizations.en.title',
       #'localizations.en.description', 'snippet.tags',
      # 'contentDetails.contentRating.ytRating'
]].copy()
```

```python
df.head(5)
```

```python
df.dtypes # list datatypes of each column
```

```python
# converting the string date to a datetime date
df['publishTime'] = pd.to_datetime(df['publishTime']) 
```

```python
# Rename our columns
df = df.rename(columns={'title':'Title',
                   'description':'Description',
                   'publishTime':'Publish_Time',
                   'duration_seconds':'Duration_Seconds',
                   'viewCount':'View_Count',
                   'likeCount':'Like_Count',
                   'commentCount':'Comment_Count'})
```

```python
df
```

```python
# Shows which column have how many null values?
df.isna().sum()
```

```python
# Check for duplicate coaster name: It show rows that have duplicate Title.
df.loc[df.duplicated(subset=['Title'])]
```

```python
# Checking an example duplicate
df.query('Title == "I Spent 50 Hours In Solitary Confinement"')
```

```python
df.columns
```

```python
df.duplicated(subset=['Title', 'Description', 'Publish_Time']).sum()
```

```python
# select just columns that are not the duplicate so just the first version.
# reset index is done in order to put index value in order
df = df.loc[~df.duplicated(subset=['Title', 'Description', 'Publish_Time'])] \
    .reset_index(drop=True).copy()
```

```python
df.head()
```

```python
df.shape
```

```python
# we removed all rows that contains nan value.
# another better approach would be replacing the nan values with some other values
df = df.dropna().reset_index(drop=True).copy()
df.head()
```

```python
df.shape
```

# Step 5: Feature Understanding
Univariate analysis: Looking each feature individually.
- Plotting Feature Distributions
    - Histogram
    - KDE
    - Boxplot

```python
df['Duration_Seconds'].value_counts()
```

```python
ax = df['Duration_Seconds'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Duractions of videos')
ax.set_xlabel('Duration_Seconds')
ax.set_ylabel('Count')
```

```python
# for continuous value like speed it would be better to plot the histogram to get better idea of distribution.
ax = df['View_Count'].plot(kind='hist',
                          bins=20,
                          title='View Count')
ax.set_xlabel('View Count')
```

```python
ax = df['View_Count'].plot(kind='kde',
                          title='Videos View Count')
ax.set_xlabel('View Count')
```

```python
# for continuous value like speed it would be better to plot the histogram to get better idea of distribution.
ax = df['Like_Count'].plot(kind='hist',
                          bins=20,
                          title='Like Count of Videos')
ax.set_xlabel('Like Count')
```

```python
ax = df['Like_Count'].plot(kind='kde',
                          title='Videos Like Count')
ax.set_xlabel('Like Count')
```

Creating new column

```python
df['like_to_view_ratio'] = df['Like_Count'] / df['View_Count']
```

```python
df
```

```python
# for continuous value like speed it would be better to plot the histogram to get better idea of distribution.
ax = df['like_to_view_ratio'].plot(kind='hist',
                          bins=20,
                          title='Like to view ratio')
ax.set_xlabel('Like to View Ratio')
```

This shows in Mr.Beast videos very few people are liking the videos who had seen it.

## Step 6: Feature Relationships
In this step, we gonna see the relationship between features.
- Scatterplot
- Heatmap Correlation
- Pairplot
- Groupby comparisons

```python
df.plot(kind='scatter',
        x='View_Count',
        y='Like_Count',
        title='View Count vs. Like Count')
plt.show()
```

```python
ax = sns.scatterplot(x='View_Count',
                y='Like_Count',
                hue='Duration_Seconds',
                data=df)
ax.set_title('View Count vs. Like Count')
plt.show()
```

```python
sns.pairplot(df,
             vars=['Duration_Seconds', 'View_Count', 'Like_Count', 'Comment_Count', 'like_to_view_ratio'])
plt.show()
```

```python
# Showing correlation between features
df_corr = df[['Duration_Seconds', 'View_Count', 'Like_Count', 'Comment_Count', 'like_to_view_ratio']].corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```
